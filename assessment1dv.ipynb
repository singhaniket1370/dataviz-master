```python
import pandas as pd
import seaborn as sns; sns.set(style="whitegrid", palette="muted")
import numpy as np
import matplotlib.pyplot as plt
```

```python
white_wine_df = pd.read_csv('winequality-white.csv', sep=";")

# Count Plot
x = sns.countplot(data=white_wine_df, x="quality")
```

```python
#Strip Plot
fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(14.5, 4.5)
fig.dpi = 300
sns.stripplot(data=white_wine_df, x="quality", y="total sulfur dioxide", jitter=True, ax=ax1)
sns.stripplot(data=white_wine_df, x="quality", y="free sulfur dioxide", jitter=True, ax=ax2)
```

```python
# Violin plot
sns.violinplot(data= white_wine_df, x="quality", y="total sulfur dioxide", scale='width', inner='quartile')
```

```python
#Strip plot
sns.stripplot(data= white_wine_df, x="quality", y="total sulfur dioxide", jitter=0.25, size=8, linewidth=.5)
```

```python
# Bar plot
sns.barplot(data=white_wine_df, x="quality", y="alcohol")
```

```python
# lmplot
sns.lmplot(data=white_wine_df, x="quality", y="alcohol")
```

```python
# Heat map
heat_table = white_wine_df[["quality", "alcohol"]].copy()
heat_table["alcohol"] = heat_table.alcohol.apply(func=lambda x: round(x * 2) / 2)
heat_table = heat_table.groupby(["quality", "alcohol"])["alcohol"].count().reset_index(name='counts')
sns.heatmap(heat_table.pivot("quality", "alcohol", "counts"))
```

```python
#Pair plot
g = sns.PairGrid(white_wine_df[["alcohol", "chlorides", "quality"]], hue="quality")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
df = pd.read_csv('winequality.csv')
```

```python
for col, value in df.items():
    if col != 'type':
        df[col] = df[col].fillna(df[col].mean())
```

```python
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(20,10))
index = 0
ax = ax.flatten()

for col, value in df.items():
    if col != 'type':
        sns.boxplot(y=col, data=df, ax=ax[index])
        index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
sns.distplot(df['free sulfur dioxide'])
```

```python
corr = df.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('winequality-red (1).csv')
print(data)
rows = data['fixedacidity;volatileacidity;citricacid;residualsugar;chlorides;freesulfurdioxide;totalsulfurdioxide;density;pH;sulphates;alcohol;quality']
```

```python
from collections import Counter
fixed_acidity = []
counter = 0
index = []
volatile_acidity = []
citric_acid = []
residual_sugar = []
chlorides = []
free_sulfur_dioxide = []
total_sulfur_dioxide = []
density = []
pH = []
sulphates = []
alcohol = []
quality = []
for value in rows:
    counter+=1
    index.append(counter)
    fixed_acidity.append(value.split(';')[0])
    volatile_acidity.append(value.split(';')[1])
    citric_acid.append(value.split(';')[2])
    residual_sugar.append(value.split(';')[3])
    chlorides.append(value.split(';')[4])
    free_sulfur_dioxide.append(value.split(';')[5])
    total_sulfur_dioxide.append(value.split(';')[6])
    density.append(value.split(';')[7])
    pH.append(value.split(';')[8])
    sulphates.append(value.split(';')[9])
    alcohol.append(value.split(';')[10])
    quality.append(value.split(';')[11])
    
# Pie chart for quality
# print(alcohol)
alcohol_count = Counter()
quality_count = Counter()



# for value in alcohol:
#     alcohol_count.update([value])
# print(alcohol_count)

for value in quality:
    quality_count.update([value])

# print(quality_count)
quality_count_Data = quality_count.values()
quality_count_Attris = quality_count.keys()

print(quality_count_Data)
print(quality_count_Attris)
```

```python
import pandas as pd
import seaborn as sns; sns.set(style="whitegrid", palette="muted")
import numpy as np
import matplotlib.pyplot as plt
```

```python
red_wine_df = pd.read_csv('winequality-red (1).csv', sep=";")
```

```python
red_wine_df.head()
```

```python
# Count Plot
x = sns.countplot(data=red_wine_df, x="quality")
```

```python
#Strip Plot
fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(14.5, 4.5)
fig.dpi = 300
sns.stripplot(data=red_wine_df, x="quality", y="totalsulfurdioxide", jitter=True, ax=ax1)
sns.stripplot(data=red_wine_df, x="quality", y="freesulfurdioxide", jitter=True, ax=ax2)
```

```python
# Violin plot
sns.violinplot(data= red_wine_df, x="quality", y="totalsulfurdioxide", scale='width', inner='quartile')
```

```python
#Strip plot
sns.stripplot(data= red_wine_df, x="quality", y="totalsulfurdioxide", jitter=0.25, size=8, linewidth=.5)
```

```python
# Bar plot
sns.barplot(data=red_wine_df, x="quality", y="alcohol")
```

```python
# lmplot
sns.lmplot(data=red_wine_df, x="quality", y="alcohol")
```

```python
# Heat map
heat_table = red_wine_df[["quality", "alcohol"]].copy()
heat_table["alcohol"] = heat_table.alcohol.apply(func=lambda x: round(x * 2) / 2)
heat_table = heat_table.groupby(["quality", "alcohol"])["alcohol"].count().reset_index(name='counts')
sns.heatmap(heat_table.pivot("quality", "alcohol", "counts"))
```

```python
#Pair plot
g = sns.PairGrid(red_wine_df[["alcohol", "chlorides", "quality"]], hue="quality")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('winequality-red (1).csv')
print(data)
rows = data['fixedacidity;volatileacidity;citricacid;residualsugar;chlorides;freesulfurdioxide;totalsulfurdioxide;density;pH;sulphates;alcohol;quality']
```

```python
from collections import Counter
fixed_acidity = []
counter = 0
index = []
volatile_acidity = []
citric_acid = []
residual_sugar = []
chlorides = []
free_sulfur_dioxide = []
total_sulfur_dioxide = []
density = []
pH = []
sulphates = []
alcohol = []
quality = []
for value in rows:
    counter+=1
    index.append(counter)
    fixed_acidity.append(value.split(';')[0])
    volatile_acidity.append(value.split(';')[1])
    citric_acid.append(value.split(';')[2])
    residual_sugar.append(value.split(';')[3])
    chlorides.append(value.split(';')[4])
    free_sulfur_dioxide.append(value.split(';')[5])
    total_sulfur_dioxide.append(value.split(';')[6])
    density.append(value.split(';')[7])
    pH.append(value.split(';')[8])
    sulphates.append(value.split(';')[9])
    alcohol.append(value.split(';')[10])
    quality.append(value.split(';')[11])
    
# Pie chart for quality
# print(alcohol)
alcohol_count = Counter()
quality_count = Counter()



# for value in alcohol:
#     alcohol_count.update([value])
# print(alcohol_count)

for value in quality:
    quality_count.update([value])

# print(quality_count)
quality_count_Data = quality_count.values()
quality_count_Attris = quality_count.keys()

print(quality_count_Data)
print(quality_count_Attris)




# for i in range(0,5):
#     alcohol_count.update(value)
# print(alcohol_count)
# Line graph for alcohol and quality
# Stack plot for alcohol and quality
# Box plot for alcohol and quality
```
